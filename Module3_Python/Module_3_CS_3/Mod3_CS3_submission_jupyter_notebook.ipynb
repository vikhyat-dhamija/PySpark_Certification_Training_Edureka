```python
#2. Name field contains full name – use regular expression to separate title, first name, last name


#Note from the above it is understood that  second column contains the Title , First name 
#First column contains the Last Name 

#Under this assumption I am separating - second column --- Title , First Name and ----first column --- Last Name 

#The main acticity of  parsing and regex is used in this case and hence we are learning the actual thing this case study wants us to learn

import re
import csv

#Task1 : Read FairDealCustomerData.csv 

with open('/mnt/home/edureka_960126/FairDealCustomerData.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        #Note title Pattern is that it starts with M and ends with . dot 
        regexp='(\S+.)\s([\S\s]+)'
        #In this 
        #row[0] is the Last Name
        #row[1] is the Title + First Name
        #row[2] is the status
        title=re.search(regexp,row[1]).group(1)
        first_name=re.search(regexp,row[1]).group(2)
        last_name=row[0]
        
        print(f'\t{title}\t{first_name}\t{last_name}\t{row[2]}')
        line_count += 1
    print(f'Processed {line_count} lines.')









```

```python
#3. Store the data in Customer Class
class customer:
   #No class variable
    
    #constructor for constructing objects with the parmeters from each line of the csv
    def __init__( self , tit , fn , ln , bl):
        #instance variable
        self.title=tit
        self.first_name=fn
        self.last_name=ln
        self.status=int(bl)
    
    
         
#array/List of customer objects , this is the table/dataset of customers

customers=[]
          
#Now with reading and parsing we are storing the data in the customer objects and then storing them in customers array

with open('/mnt/home/edureka_960126/FairDealCustomerData.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    
    for row in csv_reader:
        #Note title Pattern is that it starts with M and ends with . dot 
        regexp='(\S+.)\s([\S\s]+)'
        #In this 
        #row[0] is the Last Name
        #row[1] is the Title + First Name
        #row[2] is the status
        title=re.search(regexp,row[1]).group(1)
        first_name=re.search(regexp,row[1]).group(2)
        last_name=row[0]
        status=row[2]
        customers.append(customer(title.strip(),first_name.strip(),last_name.strip(),status.strip()))
        
        #print(f'\t{title}\t{first_name}\t{last_name}\t{row[2]}')
        line_count += 1
    print(f'Processed {line_count} lines.')
    
#checking the customers list

for c in customers :
     print(f'\t{c.title}\t{c.first_name}\t{c.last_name}\t{c.status}')





```

```python
#Task 4 : Create Custom Exception – CustomerNotAllowedException 

class CustomerNotAllowedException(Exception):
    
    def __init__( self , cust):
        self.cust=cust
        
    def __str__(self): 
        return(f'Customer:{self.cust.title}  {self.cust.first_name}  {self.cust.last_name} is not allowed to place order')
```

```python
#5. Pass a customer to function "createOrder" and throw CustomerNotAllowedException in case of blacklisted value is 1 
#create order function takes the furst name and the last name check the record of the customer in the customer list 
#And then create the order

def createOrder(f ,l):
 
    flag=0
    #checking whether the customer is allowed to place the order
    for c in customers:
        #searching for the customer
        if c.first_name== f and c.last_name == l:
            flag=1
            if(c.status == 1):
                try:
                    raise(CustomerNotAllowedException(c))

                except CustomerNotAllowedException as error: 
                    print(error)

                break
            
            else:
                print("Created the order.............")
    
    if(flag == 0):
            print("Customer Not found!!!!!")
      


```
